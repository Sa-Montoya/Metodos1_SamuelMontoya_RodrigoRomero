```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
def DerivadaProgresiva(f,x,h):
    return (1/(2*h))*(-3*f(x)+4*f(x+h)-f(x+2*h))
def DerivadaCentral(f,x,h):
    return (1/(2*h))*(f(x+h)-f(x-h))
```

```python
f = lambda x: np.sqrt(np.tan(x))
h = 0.01
a = 0.1
b = 1.1
x = np.linspace(a,b,int((b-a)/h))
y = f(x)
df_prog = DerivadaProgresiva(f,x,h)
df_central = DerivadaCentral(f,x,h)
```

$\frac{d}{dx}(\sqrt{tan(x)})=\frac{sec^2(x)}{2\sqrt{tan(x)}}$

```python
df_exact = 1/(2*np.sqrt(np.tan(x))*(np.cos(x))**2)
error_prog = abs(df_prog-df_exact)
error_central =abs(df_central-df_exact)
```

```python
fig = plt.figure(figsize=(18,6))
ax1 = fig.add_subplot(1,2,1)
ax1.scatter(x,df_prog,color='r',s=1, label='Derivada Progresiva')
ax1.scatter(x,df_central,color='b',s=1,label='Derivada Central')
ax1.scatter(x,df_exact,color='k',s=1,label='Derivada Exacta')
ax1.legend(loc=9)
plt.xlabel('x')
plt.ylabel('y')
plt.title("Derivadas")
ax1.grid(alpha=0.5)
ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x,error_prog,s=3,label='Derivada Progresiva')
ax2.scatter(x,error_central,s=3,label='Derivada Central')
plt.title("Error Nodal")
plt.xlabel('x')
plt.ylabel('Error')
ax2.legend(loc=9)
ax2.grid(alpha=0.5)
```

Teniendo en cuenta la gráfica del error nodal podemos concluir que ambas aproximaciones tienen el mismo orden de precisión en los diferentes intervalos. Por ejemplo, en x = 0.2 el error es más grande en ambos que en 0.6. Apesar de esto, la derivada central ofrace mejores resultados.
