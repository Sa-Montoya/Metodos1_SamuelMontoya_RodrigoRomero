```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
V=2
R=2
def funcion(x,y):
    return V*x*(1-((R*R)/((x*x)+(y*y))))
```

```python

N=25
x = np.linspace(-4,4,N)
y = np.linspace(-4,4,N)

h=0.001

X,Y = np.meshgrid(x,y)
```

```python
def derivativex(f,x,y,h):
    return (f(x+h,y)-f(x-h,y))/(2*h)
def derivativey(f,x,y,h):
    return (f(x,y+h)-f(x,y-h))/(2*h)
```

```python
Vx=derivativex(funcion,X,Y,h)
Vy=-derivativey(funcion,X,Y,h)

Vx = Vx.T
Vy = Vy.T
```

```python
fig1 = plt.figure(figsize=(5,5))
ax = fig1.add_subplot(1,1,1)
#circulo

r=2

for i in range(1,len(x)):
    for j in range(1,len(y)):
        if (np.sqrt((x[i]*x[i])+(y[j]*y[j])))>r:
        
            ax.quiver(x[i],y[j],Vx[i,j],Vy[i,j],color='r')

#circulo
c=plt.Circle((0,0),r,fill=False)
ax.set_aspect(1)
ax.add_artist(c)
plt.show()
```
