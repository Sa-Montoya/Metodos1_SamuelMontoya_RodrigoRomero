```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 1000
pos = np.linspace(0,1001,1001)
```

```python
def BuscarPrimo(p_0, primes):
    sig = p_0
    test = 0
    while test==0:
        sig+=1
        for i in primes:
            if (sig%i)==0:
                break
            if (i > sig/2):
                test = 1
    return sig
```

```python
def GetPrimos(N):
    cont = 0
    primos = np.array([2])
    while cont < N:
        primos = np.append(primos,BuscarPrimo(primos[cont],primos))
        cont+=1
    return primos
```

```python
primos = GetPrimos(N)
```

```python
for i in range(10):
    print(primos[i])
```

```python
fig = plt.figure( figsize=(6,6) )
ax1 = fig.add_subplot(1,1,1)
ax1.plot(pos, primos, color='g')
plt.title('NÃºmeros primos')
```
