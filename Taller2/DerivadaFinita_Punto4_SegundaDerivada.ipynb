```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
h = 0.05
x = np.linspace(-10,10,int(20/h))
f = lambda x: 1/np.sqrt(1+np.exp(-x**2))
y = f(x)
```

```python
def FirstDerivative(f,x,h):
    M = np.array([1, 0, -1])
    fx = f(x)
    D = np.sum(np.multiply(M,np.transpose(np.array([[f(x+h)],[f(x)],[f(x-h)]]))),axis=2)/(2*h)
    return D
```

$$
D^2f(x) = \frac{f(x_{j+1})-2f(x_j)+f(x_{j-i})}{h^2}
$$

$$
\mathbb{M} = \begin{bmatrix}1 & -2 & 1\end{bmatrix}
$$
$$
D^2f(x) = \frac{1}{h^2} \sum_{m=-1}^1 M[m+1]f(x_{n-m})
$$

```python
def SecondDerivative(f,x,h):
    M = np.array([1, -2, 1])
    fx = f(x)
    D = np.sum(np.multiply(M,np.transpose(np.array([[f(x+h)],[f(x)],[f(x-h)]]))),axis=2)/(h**2)
    return D
```

```python
df = FirstDerivative(f,x,h)
d2f = SecondDerivative(f,x,h)

fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x, y, color='g', label='f(x)')
ax1.plot(x,df,color='r',label='Primera derivada numérica')
ax1.plot(x,d2f,color='k',label='Segunda derivada numérica')
ax1.legend(loc=0)
ax1.grid(alpha=0.25)
```
