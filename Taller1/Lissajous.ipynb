```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 1
omega_x = 1
omega_y = 1
nsides = 5
delta = np.pi/2
```

```python
def GetLissajous(nx,ny,delta):
    theta = np.linspace(0,2*np.pi,500)
    x = A*np.sin(nx*theta)
    y = A*np.cos(ny*theta+delta)
    return x,y
```

```python
def Init(nsides):
    deltas=[0,np.pi/4,np.pi/2]
    for z in range(3):
        fig = plt.figure( figsize=(3,3) )
        delta = deltas[z]
        k=1
        for i in range(1,nsides+1):
            for j in range(i,nsides+1):
                x,y = GetLissajous(i,j,delta)
                ax = fig.add_subplot(nsides,nsides,k)
                plot = ax.plot(x,y)
                plt.axis("off")
                k += 1
            k += i

```

```python
Init(nsides)
```
