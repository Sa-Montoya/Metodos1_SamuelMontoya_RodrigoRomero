```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_0 = 0
f_1 = 1
pos = np.linspace(0,20,20)
num_aureo = np.empty(20)
num_aureo.fill((1+np.sqrt(5))/2)
fib = np.zeros(20)
fib[0] = f_0
fib[1] = f_1
```

```python
def fibonacci(f1,f2):
    f = f1 + f2
    return f
```

```python
def GetFib():
    for i in range (2,20):
        fib[i] = fibonacci(fib[i-1],fib[i-2])
```

```python
def EstimarAureo(Fib):
    est = np.zeros(19)
    est[0] = 0
    for i in range(1,19):
        est[i]=Fib[i+1]/Fib[i]
    return est
```

```python
GetFib()
est_aureo = EstimarAureo(fib)
```

```python
fig = plt.figure( figsize=(8,4) )
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)

ax1.plot(pos, fib,label='SucesiÃ³n de Fibonacci')
ax2.plot(pos[0:19],est_aureo,label="NÃºmero Aureo")
ax2.plot(pos,num_aureo,'--', label="EstimaciÃ³n del NÃºmero Aureo")
ax1.legend()
ax2.legend(loc=4)
```
