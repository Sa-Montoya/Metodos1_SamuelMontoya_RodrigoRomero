```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda x: 1/np.sqrt(1+np.exp(-x**2))
h = 0.05
```

```python
x = np.linspace(-10,10,int(20/h))
y = f(x)
```

```python
def derivative(f,x,h):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
df = derivative(f,x,h)
```

```python
df_exact = (np.exp(-x**2)*x)/((1+np.exp(-x**2))**(3/2))
```

```python
error = abs(df-df_exact)
```

```python
fig = plt.figure(figsize=(10,3))
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x, y, color='g', label='f(x)')
ax1.plot(x,df,color='r',label='Derivada numÃ©rica')
ax1.legend(loc=0)
ax1.grid(alpha=0.25)
ax2 = fig.add_subplot(1,2,2)
ax2.scatter(x, error,s=1)
ax2.grid(alpha=0.25)
ax2.title.set_text('Error de EstimaciÃ³n')
```
