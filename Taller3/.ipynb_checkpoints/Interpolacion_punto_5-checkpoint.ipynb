```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import pandas as pd
import sympy as sym
import os.path as path
import wget
```

```python
file='Data\Datos2'
url='https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/InterpolacionNewtonNoequi.csv'
if not path.exists(file):
    Path_ = wget.download(url,file)
else:
    Path_ = file
```

```python
Data = pd.read_csv(Path_,sep=',')
```

```python
Data
```

```python
X = np.float64(Data['X'])
Y = np.float64(Data['Y'])
```

```python
def NewtonGregory(X,Y,x):
    
    Sum = Y[0]
    X[1]
    Diff = np.zeros((len(X),len(Y)))
    Diff[:,0] = Y

    
            
    poly = 1.0
    
    for i in range(1,len(X)):
        h=X[i]-X[i-1]
        
       
        
        poly *= ( x - X[i-1] )
        
        for j in range(i, len(X)):
            
            Diff[j,i] = Diff[j,i-1]-Diff[j-1,i-1]
            
        Sum += poly*(Diff[i,i])/(np.math.factorial(i)*h**(i))
    
    return Sum,np.round(Diff,4)
```

```python
x = np.linspace(np.min(X),np.max(X),100)
y,_=NewtonGregory(X,Y,x)
y,_
```

```python
plt.scatter(X,Y,color='r')
plt.plot(x,y)
```

```python
x = sym.Symbol('x',Real='True')
y,_ = NewtonGregory(X,Y,x)
y
```
